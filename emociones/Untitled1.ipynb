```python
# importar librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('Emociondia.csv')
```

```python
datos.info()
```

```python
# Verificar la tabla
datos.head()
```

```python
## Grafico 1 
```

```python
# Agrupar por SEMANA y EMOCION
grupo_semana = datos.groupby(["SEMANA", "EMOCION"])["VALOR"].sum().unstack(fill_value=0)

# Graficar
grupo_semana.plot(kind="bar", figsize=(10,5))
plt.title("Semana vs Emoción vs Valor")
plt.xlabel("Semana")
plt.ylabel("Valor total")
plt.legend(title="Emoción")
plt.grid(True, linestyle='--', alpha=0.5)  # <- Aquí corregido
plt.tight_layout()
plt.show()
```

```python
# grafico 2
```

```python
# Agrupar por Día y Emoción (columna "Dia" y "EMOCION")
grupo_dia = datos.groupby(["Dia", "EMOCION"])["VALOR"].sum().unstack(fill_value=0)

# Graficar
grupo_dia.plot(kind="bar", figsize=(12,6))
plt.title("Día vs Emoción vs Valor")
plt.xlabel("Día")
plt.ylabel("Valor total")
plt.legend(title="Emoción")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
