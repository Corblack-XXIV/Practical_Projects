# Lectura de datos

```python
df = pd.read_csv('DatosSeguros.csv')
```

```python
df.info()
```

## Tratamiendo de datos

### Detectar valores vacíos o nulos

```python
df.isnull().sum()
```

### Eliminacion de valores nulos

```python
df=df.dropna()
```

```python
df.info()
```

### Eliminacion de valores duplicados

```python
df=df.drop_duplicates()
```

```python
df.info()
```

### Ordenamiento de los datos

```python
edad = df.sort_values('edad') # se mostraran los datos en orden menor a mayor
edad.head(10) # muestra los 10 primeros datos
```

```python
edad = df.sort_values('edad',ascending=False) # se mostraran los datos en orden mayor a menor
edad.head(10) # muestra los 10 primeros datos
```

### Agrupacion de datos

```python
# Definir los rangos de edades
rangos = [17,28,38,48,58,68]
```

```python
# Establecer el nombre del rango
nombrerango = ["A", "B", "C", "D", "E"]
```

```python
# Crear variable
df['Rango_Edad'] = pd.cut(df['edad'], rangos, labels = nombrerango)
```

```python
df.head(10) # muestra los rangos de edades 17-28, 2-38, 39-48, 49-58, 59-68
```

##### Gráficos de torta

###### Cálculo de distribución porcentual variable: hijos

```python
total_hijos = df['hijos'].groupby(df['hijos']).count()

etiquetas = [0,1,2,3,4,5]

colors = sns.color_palette('pastel')[0:6]

plt.pie(total_hijos, labels = etiquetas, colors = colors, autopct='%.0f%%')

plt.show()
```

###### Cálculo de distribución porcentual variable: Rango de edad

```python
total_rango_edad = df['Rango_Edad'].value_counts(sort=False)  # sort=False mantiene el orden original
	# En lugar de groupby().count(), usamos value_counts(), que es más simple y eficiente para contar categorías.
	# sort=False mantiene el orden original de las categorías.


# Usar los valores únicos de Rango_Edad como etiquetas
labels = total_rango_edad.index 
    # En lugar de usar etiquetas fijas ["A", "B", "C", "D", "E"], tomamos las etiquetas reales desde total_rango_edad.index.

# Elegir una paleta de colores adecuada
colors = sns.color_palette('Set2')[0:5]

# Crear el gráfico de pastel
plt.pie(total_rango_edad, labels=labels, colors=colors, autopct='%.1f%%', startangle=140)
plt.title("Distribución de Rango de Edad") # Agregamos un titulopara la grafica
plt.show()
```

# Analisis para el tratamiento y conclusiones

```python
# Contar la frecuencia de cada categoría en Rango_Edad
total_rango_edad = df['Rango_Edad'].value_counts(sort=False)

# Definir etiquetas correctas
labels = total_rango_edad.index  

# Definir colores con la paleta Set2
colors = sns.color_palette('Set2', n_colors=len(labels))  

# Crear el gráfico de pastel
plt.figure(figsize=(8,8))
plt.pie(total_rango_edad, labels=labels, colors=colors, autopct='%.1f%%', startangle=140)
plt.title("Distribución de Rango de Edad")
plt.show()
```

```python
# Contar la frecuencia de cada categoría en Rango_Edad
total_rango_edad = df['fumador'].value_counts(sort=False)

# Definir etiquetas correctas
labels = total_rango_edad.index  

# Definir colores con la paleta Set2
colors = sns.color_palette('Set2', n_colors=len(labels))  

# Crear el gráfico de pastel
plt.figure(figsize=(8,8))
plt.pie(total_rango_edad, labels=labels, colors=colors, autopct='%.1f%%', startangle=140)
plt.title("Distribución de Rango de Edad")
plt.show()
```

```python
# Seleccionar solo columnas numéricas
df_numerico = df.select_dtypes(include=['number'])  

# Calcular la matriz de correlación
correlacion = df_numerico.corr()

# Graficar el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```
