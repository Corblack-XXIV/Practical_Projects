```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leer el archivo de Excel en formato csv
archivo = 'base_de_datos_iris.csv'  # Cambia por la ruta real
df = pd.read_csv(archivo)
```

```python
df.head() # ver la marco de datos
```

```python
# Extraer columnas necesarias
m1 = df.iloc[:, 0:5]  # Las primeras 5 columnas
m2 = df.iloc[:, 0]    # La primera columna

# Dividir en clases (50 ejemplos por clase)
Setosa = df.iloc[1:51, 0]     # Setosa
Versicolor = df.iloc[51:101, 0]  # Versicolor
Virginica = df.iloc[101:151, 0] # Virginica

# Graficar una clase (ej. Virginica)
plt.plot(Virginica, 'go')  # 'go' = green circles
plt.title("Virginica - Primera variable")
plt.ylabel('largo sépalo')
plt.xlabel('numero de especie')
plt.grid(True)
plt.show()

# Graficar todas las clases
plt.plot(Setosa.reset_index(drop=True), 'ro', label='Setosa')
plt.plot(Versicolor.reset_index(drop=True), 'bo', label='Versicolor')
plt.plot(Virginica.reset_index(drop=True), 'go', label='Virginica')
plt.title("Clases Iris - Primera variable")
plt.ylabel('largo sépalo')
plt.xlabel('numero de especie')
plt.legend()
plt.grid(True)
plt.show()
```
